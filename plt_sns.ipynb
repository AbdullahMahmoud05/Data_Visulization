```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#3 & 4

sns.barplot(x=df['Pclass'],y=df['Fare'], data= df)
plt.title('Bar Plot Between Pclase and Fare')
plt.xlabel('Pclass')
plt.ylabel('Fare')
plt.show()

sns.boxplot(x=df['Sex'],y=df['Age'], data= df)
plt.title('Box Plot For Sex and Age')
plt.xlabel('Sex')
plt.ylabel('Age')
plt.show()

sns.histplot(df['Age'],  bins = 20 , kde = 'True')
plt.title('Histogram Displays Age')
plt.xlabel('Age')
plt.ylabel('Years')
plt.show()

sns.scatterplot(x=df['SibSp'],y=df['Age'], data= df)
plt.title('Scatter Plot For SibSp and Age')
plt.xlabel('SibSp')
plt.ylabel('Age')
plt.show()


```

```python
Numerical_df = df.select_dtypes(include=['float','int'])

corr = Numerical_df.corr()

sns.heatmap(corr, annot= True )

plt.title('Titanic Heatmap')

plt.show()
```
